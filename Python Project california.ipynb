10. create a column named total_bedroom_size

total bedrooms is 10 or less, it should be small

total bedrooms in between 11 and 1000, it should be medium

total bedrooms is more than 1000, it should be large

```python

df.loc[ df['total_bedrooms'] <= 10, 'total_bedroom_size'] = 'small' 
df.loc[ (df['total_bedrooms'] > 10) & (df['total_bedrooms'] <1000), 'total_bedroom_size' ] = 'medium' 
df.loc[df['total_bedrooms'] >1000, 'total_bedroom_size'] = 'large' 
df.head()
#using .loc function to locate the column total_bedrroms, and based on condition assigning new small,medium,large to the required condition
#all new names are added under new column total_bedrroom_size
```
